Matrix Inversion

numpy

```python
import numpy as np
```

```python
#buat matrix X ukuran 2x2 menggunakan numpy
X = np.array([[4, 2], [-5, -3]])
print(f"Matriks  X adalah: \n{X}")
```

```python
#buat invers dari matrix X dan beri nama Xinv menggunakan numpy
Xinv = np.linalg.inv(X)
print(f"Invers matriks X adalah: \n{Xinv}")
```

```python
#kalikan X dengan Xinv menggunakan numpy
print(f"hasil perkalian adalah: \n{np.dot(Xinv, X)}")
```

4b + 2x = 4

-5b - 3 = -7

```python
X = np.array([[4, 2], [-5, -3]])
print(f"Tampilkan: \n{X}")
```

```python
V = ([4, -7])
print(f"Tampilkan: \n{V}")
```

```python
C =  np.linalg.solve(X, V)
print(f"Hasil: \n{C}")
```

torch dan tensorflow

```python
import torch
import tensorflow as tf
```

```python
#torch
X_pt = torch.tensor([[4, 2], [-5, -3]])
print(f"Matriks X: \n{X_pt}")
```

```python
V_pt  = torch.tensor([4, -7])
print(f"Vektor V: \n{V_pt}")
```

```python
C_pt  = torch.linalg.solve(X_pt.float(), V_pt.float())
print(f"Hasil: \n{C_pt}")
```

```python
#tensorflow

X_tf = tf.Variable([[4., 2.], [-5., -3.]])
print(f"Matriks X: \n{X_tf}")
```

```python
V_tf = tf.Variable([[4.],[7.]])
print(f"Vektor V: \n{V_tf}")
```

```python
C_tf = tf.linalg.solve(X_tf, V_tf)
print(f"Hasil: \n{C_tf}")
```
