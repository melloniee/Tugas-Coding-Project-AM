numpy

```python
import numpy as np
```

```python
#1. buat matrix A dengan ukuran 2x3 dan vector b berukuran 3x1 menggunakan numpy
A = np.array([[4, 3, 4], [4, 5, 7]])
print(f"Matrix A adalah: \n{A}")
```

```python
#vector 3x1
b = np.array([[9], [4], [5]])
print(f"Vector b adalah: \n{b}")
```

```python
#t2. ampilkan ukuran matrix A dan vector b

print(f"Ukuan matrix A adalah: {A.shape}")
print(f"Ukuran vector b adalah: {b.shape}")
```

```python
#3. kalikan matrix A dengan vector b
print(f"hasil perkalian matriks A dan vector b adalah: \n{np.dot(A, b)}")
```

pytorch

```python
import torch
```

```python
#1. buat matrix A_pt dengan ukuran 2x3 dan vector b_pt berukuran 3x1 menggunakan pytorch
A_pt = torch.tensor([[1, 5, 8], [4, 6, 4]])
print(f"Matrix A_pt adalah: \n{A_pt}")
```

```python
#vector 3x1
b_pt = torch.tensor([[4], [3], [7]])
b_pt
```

```python
#3. kalikan matrix A_pt dengan vector b_pt
print(f"hasil perkalian matrix A_pt dengan vector b_pt adalah: \n{torch.matmul(A_pt, b_pt)}")
```

tensorflow

```python
import tensorflow as tf
```

```python
#1. buat matrix A_tf dengan ukuran 2x3 dan vector b_tf berukuran 3x1 menggunakan tensorflow
A_tf = tf.Variable([[1, 5, 7], [4, 6, 9]])
print(f"Matrix A_tf adalah: \n{A_tf}")
```

```python
#vector 3x1
b_tf = tf.Variable([[7], [8], [5]])
print(f"Vector b_tf adalah: \n{b_tf}")
```

```python
#3. kalikan matrix A_tf dengan vector b_tf
print(f"hasil perkalian matrix A_tf dgn vector b_tf adalah: \n{tf.linalg.matmul(A_tf, b_tf)}")
```
