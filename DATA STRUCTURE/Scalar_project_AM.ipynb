TENSORFLOW

```python
#lakukan tahapan pada soal 1 dengan menggunakan tensorflow
#buat variabel skalar x dan simpan suatu nilai skalar

import tensorflow as tf

x_tf = tf.Variable(70)
print(f"nilai dari  x_tf adalah: {x_tf}")
```

```python
# tampilkan tipe data dari x

print(f"tipe data dari x_tf adalah: {x_tf.dtype}")
```

```python
# buat variabel skalar y dan simpan suatu nilai skalar

y_tf  = tf.Variable(80)
print(f"nilai dari y_tf adalah: {y_tf}")
```

```python
# tampilkan tipe data dari y

print(f"tipe data dari y_tf adalah: {y_tf.dtype}")
```

```python
# jumlahkan x+y

sum_tf  = x_tf +  y_tf
print(f"hasil penjumlahan dari x_tf dan y_tf adalah: {sum_tf}")
```

```python
#buat variabel bernama x_float dan simpan suatu nilai desimal

x_float_tf = tf.Variable(0.57)
print(f"nilai dari x_float_tf adalah: {x_float_tf}")
```

```python
#jumlahkan x_float dengan y

sum_float = x_float_tf + tf.cast(y_tf, tf.float32)
print(f"hasil penjumlahan dari x_float_tf dan y_tf adalah: {sum_float}")
```

```python
#tampilkan tipe data dari x_float

print(f"tipe data dari x_float_tf adalah: {x_float_tf.dtype}")
```
