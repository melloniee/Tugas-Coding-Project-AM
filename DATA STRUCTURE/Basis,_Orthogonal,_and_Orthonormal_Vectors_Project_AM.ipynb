```python
import numpy as np
```

```python
# buat vector i dengan elemen (1,0)
i = np.array([1, 0])
print(f"vector i adalah: \n{i}")
```

```python
# buat vector j dengan elemen (0,1)
j = np.array([0, 1])
print(f"vector j adalah: \n{j}")
```

```python
# buktikan bahwa vector i dan vector j itu orthogonal
print(f"hasil dari vector i dan j adalah: {np.dot(i, j)}")
```

```python
# Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1, 2])
w = np.array([2, -1])
print(f"hasil dari v dan w adalah: {np.dot(v, w)}")
```
