```python
# 1. buat vector x dengan ukuran 3x1

import numpy as np
x = np.array([[35, 30, 80]])
print(f"vektor x adalah: {x}")
```

```python
# hitung nilai L2 norm dari vector x

norm = (35**2 + 30**2 + 80**2)**(1/2)
print(f"hasil L2 norm: {norm}")
```

```python
#2. buat vector v dengan elemen (3,4)

v = np.array([3, 4])
print(f"tampilkan vektor v: {v}")
```

```python
# hitung nilai L2 norm dari vector x

norm_v = (3**2 + 4**2)**(1/2)
print(f"hasil L2 norm: {norm_v}")
```

```python
#tampilkan unit vector dari vector v

unit_v = v /norm_v
print(f"unit vector dari vector v adalah: {unit_v}")
```

```python
# hitung panjang unit vector dari vector v

p_u_v = np.linalg.norm(unit_v)
print(f"hasil panjang unit vector dari vector v adalah: {p_u_v}")
```
