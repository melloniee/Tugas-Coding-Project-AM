```python
# IMPORT LIBRARY
import numpy as np
import torch
import tensorflow as tf
```

```python
# 1. Membuat Matrix 2x3
X = np.array([[1, 2, 3],
              [4, 5, 6]])

X_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]])

X_tf = tf.Variable([[1, 2, 3],
                    [4, 5, 6]])
```

```python
# 2. Penjumlahan semua elemen
print("NumPy jumlah:", X.sum())
print("PyTorch jumlah:", torch.sum(X_pt))
print("TensorFlow jumlah:", tf.reduce_sum(X_tf))
```

```python
# 3. Penjumlahan berdasarkan ROW
print("NumPy jumlah baris:", X.sum(axis=1))
print("PyTorch jumlah baris:", torch.sum(X_pt, dim=1))
print("TensorFlow jumlah baris:", tf.reduce_sum(X_tf, axis=1))
```

```python
 #4. Penjumlahan berdasarkan KOLOM
print("NumPy jumlah kolom:", X.sum(axis=0))
print("PyTorch jumlah kolom:", torch.sum(X_pt, dim=0))
print("TensorFlow jumlah kolom:", tf.reduce_sum(X_tf, axis=0))
```

```python
# 5. MIN, MAX, MEAN
# MIN
print("NumPy minimum:", X.min())
print("PyTorch minimum:", torch.min(X_pt))
print("TensorFlow minimum:", tf.reduce_min(X_tf))

# MAX
print("NumPy maksimum:", X.max())
print("PyTorch maksimum:", torch.max(X_pt))
print("TensorFlow maksimum:", tf.reduce_max(X_tf))

# MEAN
print("NumPy rata-rata:", X.mean())
print("PyTorch rata-rata:", torch.mean(X_pt.float()))
print("TensorFlow rata-rata:", tf.reduce_mean(tf.cast(X_tf, tf.float32)))
```

```python
# 6. Perkalian semua elemen
print("NumPy hasil perkalian:", X.prod())
print("PyTorch hasil perkalian:", torch.prod(X_pt))
print("TensorFlow hasil perkalian:", tf.reduce_prod(X_tf))
```
