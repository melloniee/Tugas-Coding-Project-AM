VISUALIZING LINEAR SYSTEM

y = 2x + 1

3x - y = 5

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 400)
```

```python
y1 = (2 * x) + 1
```

```python
y2 = (3 * x) - 5
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.xlabel('y')
ax.set_xlim([0, 20])
ax.set_ylim([0, 20])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='yellow')
plt.axvline(x = 6, color='purple', linestyle='--')
plt.axhline(y = 13, color="red", linestyle='--')

```
