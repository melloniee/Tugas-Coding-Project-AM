```python
#Import Library
import numpy as np
import torch
import tensorflow as tf
```

```python
# 1. Membuat vector 1x3
# NumPy
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print(f"x: {x}")
print(f"y: {y}")

# PyTorch
x_pt = torch.tensor([1, 2, 3])
y_pt = torch.tensor([4, 5, 6])

print(f"x_pt: {x_pt}")
print(f"y_pt: {y_pt}")

# TensorFlow
x_tf = tf.Variable([1, 2, 3])
y_tf = tf.Variable([4, 5, 6])

print(f"x_tf: {x_tf}")
print(f"y_tf: {y_tf}")
```

```python
# 2. Dot Product
# NumPy
print("NumPy dot:", np.dot(x, y))
# PyTorch
print("PyTorch dot:", torch.dot(x_pt, y_pt))
# TensorFlow
print("TensorFlow dot:", tf.tensordot(x_tf, y_tf, axes=1))
```

```python
#lakukan perhitungan excercise menggunakan collab
#1
Y = np.array([[42, 4, 7, 99], [-99, -3, 12, 22]])
print(f"tampilkan \n{Y}")

#y transpose
Y_t = Y.T
print(f"hasil tranpose matriks Y adalah: \n{Y_t}")
```

```python
#2

A = np.array([[25, 10], [-2 , 1]])
B = np.array([[-1, 7], [10,  8]])
print(f"tampilkan \n{A}")
print(f"tampilkan \n{B}")

#hadddamart

result = A * B
print(f"hasil perkalian haddamard matriks diatas adalah: \n{result}")
```

```python
#3

w = np.array([-1, 2, -2])
x = np.array([5, 10, 0])
print(f"tampilkan w: {w}")
print(f"tampilkan x: {x}")

#dot product
result = np.dot(w, x)
print(f"hasil dot product matriks diatas adalah: {result}")
```
