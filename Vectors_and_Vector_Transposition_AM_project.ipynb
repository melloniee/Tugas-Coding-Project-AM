```python
#1.  buat vector x dengan ukuran 3x1

import numpy as  np
x = np.array([[20],
              [35],
              [45]])
print(f"menampilkan vector x:  \n{x}")
```

```python
#hitung panjang vector x

print(f"panjang vektor x adalah: {len(x)}")
```

```python
#tampilkan ukuran vector x

print(f"ukuran dar vector x adalah: {x.shape}")
```

```python
#tampilkan tipe data dari vector x

print(f"tipe data dari vector x adalah: {type(x)}")
```

```python
# akses elemen pertama dari vector x

print(f"elemen pertama dari vector x adalah: {x[0]}")
```

```python
##tampilkan tipe data elemen pertama dari vector x

print(f"tipe data semua elemen adalah: {type(x[0,0])}")
```

2. Soal nomor 2

```python
#2. lakukan transpose pada vector x dan simpan pada variabel x_t

x_t = x.T
print(f"hasil transpose x: {x_t}")
```

```python
# tampilkan ukuran dari vector x_t

print(f"ukuran dari vector x_t adalah: {x_t.shape}")
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t

y = np.array([[21, 30, 45]])
print(f"vector y adalah: {y}")

y_t = y.T
print(f"hasil transpose y: \n{y_t}")
```

```python
# tampilkan ukuran dari vector x_t

print(f"ukuran dari vector y_t adalah: {y_t.shape}")
```

3.  lakukan soal nomor 2 menggunakan pytorch dan tensorflow

PyTorch

```python
import torch


#lakukan transpose pada vector x dan simpan pada variabel x_t

x_pt = torch.tensor(x)
x_pt_t = x_pt.T
print(f"hasil transpose x: {x_pt_t}")
```

```python
# tampilkan ukuran dari vector x_t

print(f"ukuran dari vector x adalah: {x_pt_t.shape}")
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t

y_pt = torch.tensor(y)
y_pt_t = y_pt.T
print(f"hasil transpose y: \n{y_pt_t}")
```

```python
# tampilkan ukuran dari vector x_t

print(f"ukuran dari vector  x adalah: {x_pt_t.shape}")
```

TensorFlow

```python
import tensorflow as tf

#lakukan transpose pada vector x dan simpan pada variabel x_t

x_tf = tf.transpose(x)
print(f"hasil dari transpose x: {x_tf}")
```

```python
#tampilkan ukuran dari vector x_tf

print(f"ukuran dari vector x_tf adalah: {x_tf.shape}")
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t

y_tf = tf.transpose(y)
print(f"hasil dari transpose y_tf : \n{y_tf}")
```

```python
#tampilkan ukuran dari vector x_t

print(f"ukuran dari vector x_tf adalah: {x_tf.shape}")
```
